```python
import requests
import json
import pandas as pd


class Scrappy():    
    def __init__(self):   
        self.url = "https://api.crunchbase.com/v3.1/organizations/pest "
        self.webpage = requests.get(self.url).json()["data"]    
        
    def to_DataFrame(self):
        self.df = pd.DataFrame.from_dict(self.webpage)
        
    def show(self):
        return self.df
```

```python
w1 = Scrappy()
```

```python
w1.to_DataFrame()
```

```python
w1.show()
```
