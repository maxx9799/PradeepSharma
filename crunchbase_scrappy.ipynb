```python
from selenium import webdriver
import pandas as pd
import time

driver = webdriver.Chrome(r'E:\chrome\chromedriver.exe')
url = 'https://www.crunchbase.com/organization/agraquest'
driver.get(url)
```

```python
Organization_Name = []
name1 = driver.find_element_by_class_name('profile-name')
#print(name1.text)
Organization_Name.append(name1.text)

descc = driver.find_elements_by_css_selector('button.mat-focus-indicator.mat-button.mat-button-base.mat-accent.ng-star-inserted')
if len(descc):
    descc[0].click()
time.sleep(1)

ul = driver.find_elements_by_css_selector('div.expanded')
Full_Description =[]
if len(ul):
    Full_Description.append(ul[0].text)

Transaction_Name=[]
Transaction_Name_URL=[]
next1 = driver.find_elements_by_css_selector('a.component--field-formatter.field-type-identifier.link-accent.ng-star-inserted')

Transaction_Name.append(next1[3].text)
Transaction_Name_URL.append(next1[3].get_attribute('href'))

elem = driver.find_elements_by_css_selector('ul.text_and_value')

Acquired_by=[]
Announced_Date=[]
Price=[]
for i in range(len(elem)):
                for word in range(len(elem[i].text.split())):
                    if elem[i].text.split()[word] == 'Acquired':
                        if elem[i].text.split()[word+1] == 'by':
                            Acquired_by.append(elem[i].text.split()[word+2])
                    if elem[i].text.split()[word] == 'Announced':
                        if elem[i].text.split()[word+1] == 'Date':
                            Announced_Date.append(elem[i].text.split()[word+2])
                            Announced_Date.append(elem[i].text.split()[word+3])
                            Announced_Date.append(elem[i].text.split()[word+4])
                    if elem[i].text.split()[word] == 'Price':
                            Price.append(elem[i].text.split()[word+1])

next1 = driver.find_elements_by_css_selector('a.mat-tab-link.mat-focus-indicator.ng-star-inserted')
for k in next1:
    if k.text == 'Technology':
        k.click()
        time.sleep(1)

    elem = driver.find_elements_by_css_selector('markup-block')

    IPqwery_Patents_Granted = []
    IPqwery_Trademarks_Registered = []
    IPqwery_Most_Popular_Patent_Class = []
    Websites_rank =[]
    for i in range(len(elem)):
                for word in range(len(elem[i].text.split())):
                    if elem[i].text.split()[word] == 'registered':
                        if elem[i].text.split()[word+1] == 'patents':
                            IPqwery_Patents_Granted.append(elem[i].text.split()[word-1])
                    if elem[i].text.split()[word] == 'trademarks':
                            IPqwery_Trademarks_Registered.append(elem[i].text.split()[word-1])
                    if elem[i].text.split()[word] == 'websites':
                        if elem[i].text.split()[word+1] == 'globally':
                            Websites_rank.append(elem[i].text.split()[word-2])
                    if elem[i].text.split()[word] == 'most':
                        if elem[i].text.split()[word+1] == 'popular':
                                if elem[i].text.split()[word+2] == 'class':
                                    IPqwery_Most_Popular_Patent_Class.append(elem[i].text.split()[word+4:-3])
if len(IPqwery_Most_Popular_Patent_Class):
    IPqwery_Most_Popular_Patent_Class[0] = " ".join(IPqwery_Most_Popular_Patent_Class[0])

final = [Organization_Name,
Full_Description,
IPqwery_Patents_Granted,
IPqwery_Trademarks_Registered,
IPqwery_Most_Popular_Patent_Class,
Websites_rank,Transaction_Name,
Transaction_Name_URL,
Acquired_by,
Announced_Date,
Price]

for i in final:
    if not len(i):
        i.append("None")
```

```python
df = pd.DataFrame(list(zip(final[0],final[1],final[2],final[3],final[4],final[5],final[6],final[7],final[8],final[9],final[10])),
columns =['Organization_Name',
'Full_Description',
'IPqwery_Patents_Granted',
'IPqwery_Trademarks_Registered',
'IPqwery_Most_Popular_Patent_Class',
'Websites_rank','Transaction_Name',
'Transaction_Name_URL',
'Acquired_by',
'Announced_Date',
'Price'])
```
